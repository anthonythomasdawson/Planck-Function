The following code:

Firstly prints the value of $U(f,T)$ for a user inputted frequency

Secondly gives a plot of Rayleigh-Jeans law over a range of frequncies to show that this leads to the Ultraviolet catastrophe. Using the formula $$U(f, T) = \frac{8 \pi f^{2}}{c^3}{kT}$$

Thirdly the program gives a plot of Planck's function for Temperatures of 2000K, 3000K, 4000K, 5000K, 6000K alongside Rayleigh-Jeans plot

Fourthly the program finds the peak of the Planck function for each temperature and its corresponding frequency. Using this the function then calculates the wavelength at this peak and uses it to estimate the constant in Wien's displacement law

Finally the programme performs integrals of the Planck function for temperatures of 2000K, 4000K, and 6000K using trapezoidal and Simpson methods. The programme then evaluates if these are proportional to $T^4$ from Stefan Boltzmanns law $ I = \sigma T^4$

```python
#We will need to use NumPy
import numpy as np

#Set the constants
k = 1.38e-23                                  #Initialise Boltzmann constant k in J K ^-1
c = 3.0e8                                     #Initial velocity in m/s


#Input the frequency in Hertz
f = eval(input("enter value for frequency in Hertz:"))
#Input the Temperature in Kelvin
T = eval(input("enter value for Temperature in Kelvin:"))


#calculate the value of U(v,T)
U=(8 * np.pi * (f ** 2) * k * T)/(c ** 3)

#output the answer in scientific notation
print("the spectral radiance is: {0:10.2e} W/m^2".format(U))
```

Task 5

Will use Trapezoidal and Simpson integration methods to estimate the area beneath the curve for above plots at temperatures of 2000K, 4000K, and 6000K.

The trapezoidal rule works by dividing the region up into a number of rectangles, and to give an improved estimate, a triangle is added to each rectangle giving a closer estimate to the true area under the curve

Insead of a linear approximation (making the triangles) in the trapezoid rule, the Simpson Rule instead makes a quadratic approximation of the "leftovers" in the rectangle rule.

```python
import numpy as np
from scipy.integrate import simps
from numpy import trapz

y = U

# Compute the area using the composite trapezoidal rule.
area = trapz(y, dx=10000)
print("For T=2000K Trapezoidal Rule Estimate of Area is:{0:10.4e}".format(area))

# Compute the area using the composite Simpson's rule.
area = simps(y, dx=10000)
print("For T=2000K Simpson Rule Estimate of Area is:{0:10.4e}".format(area))

#Divide calculated area by T^4 to check if proportional
ratio = area/2000**4
print("Ratio check if proportional is:{0:10.4e}".format(ratio))


y = U_3

# Compute the area using the composite trapezoidal rule.
area = trapz(y, dx=10000)
print("For T=4000K Trapezoidal Rule Estimate of Area is:{0:10.4e}".format(area))

# Compute the area using the composite Simpson's rule.
area = simps(y, dx=10000)
print("For T=4000K Simpson Rule Estimate of Area is:{0:10.4e}".format(area))

#Divide calculated area by T^4 to check if proportional
ratio = area/4000**4
print("Ratio check if proportional is:{0:10.4e}".format(ratio))


y = U_5

# Compute the area using the composite trapezoidal rule.
area = trapz(y, dx=10000)
print("For T=6000K Trapezoidal Rule Estimate of Area is:{0:10.4e}".format(area))

# Compute the area using the composite Simpson's rule.
area = simps(y, dx=10000)
print("For T=6000K Simpson Rule Estimate of Area is:{0:10.4e}".format(area))

#Divide calculated area by T^4 to check if proportional
ratio = area/6000**4
print("Ratio check if proportional is:{0:10.4e}".format(ratio))
```

When each calculated value of area was divided by respective $T^4$, a constant of approximately 3.6e-25 was found. This demonstrates that the calculated areas are proportional to $T^4$ as in stefan boltzmann law.

There is a slight discrepency between trapezoidal and simpson rule estimates. The trapezoidal rule consistently underestimates the area under the curve comapred to the simpson rule. This is because the triangular area estimates (error step proportional to h^2) do not fit the curves as well as the quadratic estimates performed in the simpson rule (error step proportional to h^4), and therefore the simpson rule estimates will be more accurate.
